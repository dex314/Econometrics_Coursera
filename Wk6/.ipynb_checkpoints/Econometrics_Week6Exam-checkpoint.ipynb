#### PART B

```python
yb = pd.DataFrame(y)
yb['d_y'] = y.diff()
yb['lag1'] = y.shift(1)
yb['d_lag1'] = y.diff().shift(1)
yb['d_lag2'] = y.diff().shift(2)
yb['d_lag3'] = y.diff().shift(3)

yb.dropna(inplace=True)
yb.drop(['TOYOTA_SA'],axis=1,inplace=True)
yadf = yb['d_y']
xadf = yb.drop(['d_y'], axis=1)

modb = sma.OLS(endog=yadf, exog=sma.add_constant(xadf))
fitb = modb.fit()
print fitb.summary2()
```

```python
y_adf = smts.adfuller(y, maxlag=3, autolag=None, regression='c', regresults=True) #c for constant only, no trend
x_adf = smts.adfuller(x, maxlag=3, autolag=None, regression='c', regresults=True)
fig, ax=subplots(1,2,figsize=(12,4))
ax = ax.ravel()
ax[0].plot(y_adf[0], 'ks', label='ADF Value', ms=5)
ax[0].plot(y_adf[2]['1%'], 'rs', ms=5, label='99th Percentile')
ax[0].plot(y_adf[2]['5%'], 'gs', ms=5, label='95th Percentile')
ax[0].plot(y_adf[2]['10%'], 'bs', ms=5, label='90th Percentile')
ax[0].set_title('Augmented Dickey Fuller Test for Y - Toyota SA')
ax[0].legend()
ax[1].plot(x_adf[0], 'ks', label='ADF Value', ms=5)
ax[1].plot(x_adf[2]['1%'], 'rs', ms=5, label='99th Percentile')
ax[1].plot(x_adf[2]['5%'], 'gs', ms=5, label='95th Percentile')
ax[1].plot(x_adf[2]['10%'], 'bs', ms=5, label='90th Percentile')
ax[1].set_title('Augmented Dickey Fuller Test for X - Other SA')
ax[1].legend()

print "Y ADF :\n"
print y_adf
print "\n \n X ADF: \n"
print x_adf, "\n"
```

```python
yADF_Results = y_adf[3]
print yADF_Results.resols.summary2(xname=['ylag1','ydiff_lag1','ydiff_lag2','ydiff_lag3','const'])
```

__(b)__:

__The value for $y_{t-1}$ is -0.0832 and the t-value is significant at the 5% level with a value of 0.0246.__


__If we manually calculate the OLS regression using the differenced y and its subsequent lags and compare it to the ADF test regression results from statsmodels, we can see that they are nearly identitical. Thus we can use the statsmodels test as set up above and:__

__(i)__:
__The Augmented Dickey Fuller test for y has a test statistic of -2.2623 and we cannot reject the Null Hypothesis that there is a unit root and the data is not stationary. This can also be confirmed based on the plot above and the p-value of the test statistic.__ 


__(ii)__: 
__The ADF test for x has a test statistic of -2.1057 and we cannot reject the Null Hypothesis that there is a unit root and the data is not stationary. This can also be confirmed based on the plot above and the p-value of the test statistic.__


__Additionally, based on the Autocorrelation plots below, we have further confirmation.__

```python
fig1, ax1=subplots(1,2,figsize=(12,4))
ax1.ravel()
y_pacf = sma.tsa.graphics.plot_acf(y.dropna(), lags=20, ax=ax1[0]); ax1[0].set_xlabel('y_acf')
x_pacf = sma.tsa.graphics.plot_acf(x.dropna(), lags=20, ax=ax1[1]); ax1[1].set_xlabel('x_acf')
fig2, ax2=subplots(1,2,figsize=(12,4))
ax2.ravel()
y_pacf = sma.tsa.graphics.plot_acf(y.diff().dropna(), lags=20, ax=ax2[0]); ax2[0].set_xlabel('diff_y_pacf')
x_pacf = sma.tsa.graphics.plot_acf(x.diff().dropna(), lags=20, ax=ax2[1]); ax2[1].set_xlabel('diff_x_pacf')
```

#### PART C

```python
## STEP 1:
modc_1 = sma.OLS(y, sma.add_constant(x))
fitc_1 = modc_1.fit()
res = fitc_1.resid
```

```python
## STEP 2:
modc_2 = smts.adfuller(res, maxlag=3, autolag=None, regression='c', regresults=True) #c for constant only, no trend
resultsC = modc_2[3]
print resultsC.resols.summary2(xname=['e_lag1','d_e_lag1','d_e_lag2','d_e_lag3','const'])
print "\n"
print "ADF RESULTS ON THE RESIDUAL OF X AND Y \n"
print modc_2
```

__(c)__:

__Based on the residual in part two, and the results of the ADF test, it appears we can reject the Null Hypothesis that the residuals are not stationary, thus it is not cointegrated between X and y and the linear combination of the two can form a stationary series.__

__The ACF and PACF plots are created above. However many more lagged terms than "1 to 5, 10, and 12" appear to be significant. We will just go with the ones listed in the instructions of the assignment since handling this selection should essentially take of the rest.__

__As an aside, I tried to use the _significance test_ from the lecture slides just incase the statsmodels methods were doing something slightly different but this still did not give the expected lag structure. The results can be seen above the ACF and PACF plots.__

```python
yd = pd.DataFrame(y)
yd['d_y'] = y.diff()
# yd['lag1'] = y.shift(1)
for i in np.array([1,2,3,4,5,10,12]):
    yd['dlag'+str(i)] = y.diff().shift(i)

yd.dropna(inplace=True)
yd.drop(['TOYOTA_SA'],axis=1,inplace=True)
yd_adf = yd['d_y']
xd_adf = yd.drop(['d_y'], axis=1)

modD = sma.OLS(endog=yd_adf, exog=sma.add_constant(xd_adf))
fitD = modD.fit()
print fitD.summary2()

figchk, axchk=subplots(1,1)
chk = sma.tsa.graphics.plot_pacf(fitD.resid, lags=20, ax=axchk)
xlabel('PACF of Residuals')
```

__Examining the PACF of the residuals, it can be seen that the model created does cover the majority of the auto-correlation lag structure with the exception of the 18th lag. The model also highlights (based on the p-values) that all of the lag structures added are statsitically significant from a standpoint of inclusion and based on the the calculated t-value.__

#### PART E

```python
## ADD IN ERROR TERM
ye = yd.copy()
ye['ECM'] = (y - 0.45*x).shift()

ye.dropna(inplace=True)
# ye.drop(['TOYOTA_SA'],axis=1,inplace=True)
ye_adf = ye['d_y']
xe_adf = ye.drop(['d_y'], axis=1)

modE = sma.OLS(endog=ye_adf, exog=sma.add_constant(xe_adf))
fitE = modE.fit()
print fitE.summary2()
```

__(e):__

__According to the statsmodels analysis above, at a value of _0.0319_, the Error Correction Term (ECM) is significant at a 5% level but not at a 1% level. For this to have been the case it would have had to have been less than 0.01.__

```python
## NEED TO START WITH THE WHOLE DATA FRAME OTHER WISE
## WHEN YOU DROPNA() YOU LOSE ALL OF YOUR OBERVATIONS
yt2 = df['TOYOTA_SA']
ydT = pd.DataFrame(yt2, columns=['TOYOTA_SA'])
ydT['Const'] = 1
ydT['d_y'] = yt2.diff()
# ydT['lag1'] = yt2.shift(1)
for i in np.array([1,2,3,4,5,10,12]):
    ydT['dlag'+str(i)] = yt2.diff().shift(i)


ydT.dropna(inplace=True)
ydT.drop(['TOYOTA_SA','d_y'],axis=1,inplace=True)

forecastD = pd.DataFrame(fitD.predict(ydT), index=ydT.index)
dataD = forecastD[forecastD.index<='1999-12-01']; dataD.columns=['Actuals']
forcD = forecastD[forecastD.index>'1999-12-01']; forcD.columns=['Forecast D']
```

```python
## NEED TO START WITH THE WHOLE DATA FRAME OTHER WISE
## WHEN YOU DROPNA() YOU LOSE ALL OF YOUR OBERVATIONS
xt2 = df['OTHER_SA']
yeT = ydT.copy()
yeT['EMC'] = (yt2 - 0.45*xt2).shift()

yeT.dropna(inplace=True)

forecastE = pd.DataFrame(fitE.predict(yeT), index=yeT.index)
dataE = forecastE[forecastE.index<='1999-12-01']; dataE.columns=['Actuals']
forcE = forecastE[forecastE.index>'1999-12-01']; forcE.columns=['Forecast E']
```
